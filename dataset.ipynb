```python
from dataset import prepare_dataloader
import matplotlib.pyplot as plt
import numpy as np

dataset = prepare_dataloader('data_depth_annotated/train', [], False, 16, (256, 512), 1, shuffle=True)

n_instances, loader = dataset

fig, ax = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(23, 8)

for batchidx, instance in enumerate(loader):
    i0, i1, d0, d1 = instance
    ax[0][0].imshow(np.moveaxis(i0[0].numpy(), 0, -1))
    ax[0][1].imshow(np.moveaxis(i1[0].numpy(), 0, -1))
    ax[1][0].imshow(np.moveaxis(d0[0].numpy(), 0, -1), cmap='plasma')
    ax[1][1].imshow(np.moveaxis(d1[0].numpy(), 0, -1), cmap='plasma')

    break
    
```
